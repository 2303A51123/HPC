```python
import threading
import time

barrier = threading.Barrier(3)

def worker(n):
    print("Thread", n, "before barrier")
    time.sleep(1)
    barrier.wait()
    print("Thread", n, "after barrier")

for i in range(10):
    threading.Thread(target=worker, args=(i,)).start()
```

```python
import threading

lock = threading.Lock()
value = 0

def add():
    global value
    for _ in range(100):
        with lock:   # atomic-like update
            value += 1
```

```python
from concurrent.futures import ThreadPoolExecutor

nums = [1,2,3,4,5,6,7,8]

def square(x):
    return x*x

with ThreadPoolExecutor() as ex:
    result = sum(ex.map(square, nums))  # reduction sum

print(result)
```

```python
import threading

lock = threading.Lock()

def task(i):
    with lock:
        print("Ordered output:", i)

for i in range(5):
    threading.Thread(target=task, args=(i,)).start()
```

```python
import threading

once_lock = threading.Lock()
done = False

def task():
    global done
    with once_lock:
        if not done:
            print("Single section executed")
            done = True
```
